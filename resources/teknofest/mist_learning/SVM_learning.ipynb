```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
import sklearn.svm as svm
from sklearn.metrics import f1_score, recall_score, precision_score, roc_auc_score, roc_curve, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import pickle
from joblib import dump, load
```

```python

ls = []
with open('geometry.txt', 'r') as file:
    for line in file:
        if line[0] != 'i':
            ls.append(list(map(float, line.strip().split(' '))))
```

```python
data = np.asarray(ls)
```

```python
X = data[:, 0:2]
y = data[:, 2]

pos = np.where(y == 1)
neg = np.where(y == 0)
plt.scatter(X[pos, 0], X[pos, 1], marker='o', c='b', label = 'Visible')
plt.scatter(X[neg, 0], X[neg, 1], marker='x', c='r', label = 'Not visible')
plt.xlabel('Mist density')
plt.ylabel('L2 Distance')
plt.legend()
```

```python
(trainData, testData, trainLabels, testLabels) = train_test_split(data[:, :2], data[:, 2:], test_size=0.3, random_state=41)
```

```python
data
```

```python
sc = StandardScaler()
trainData = sc.fit_transform(trainData)
testData = sc.transform(testData)
```

```python
dump(sc, '../StandartScalerSVC_teknofest.bin', compress=True)
```

```python
params = {'kernel':('linear', 'rbf', 'sigmoid'), 'C': np.linspace(0.1, 30, 35)}
model_svc = svm.SVC(random_state=41)
```

```python
finder_svc = GridSearchCV(model_svc, params, scoring='f1_micro')
finder_svc.fit(trainData, trainLabels)
```

```python
svc_res_df = pd.DataFrame(finder_svc.cv_results_).sort_values(by=['mean_test_score'], ascending=False)
svc_res_df.head()
```

```python
preds = finder_svc.predict(testData)
```

```python
f1_score(preds, testLabels)
```

```python
filename = '../SVM_mist_classifier_teknofest.sav'
pickle.dump(finder_svc, open(filename, 'wb'))
```
